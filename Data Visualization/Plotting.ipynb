```julia
year = [x for x in range(2000, length= 10)]
```

```julia
sales = 100*rand(10)
```

```julia
plot(year, sales, labels="Line Plot")
```

```julia
scatter!(year, sales, labels="Scatter Plot")
```

```julia
xlabel!("Year")
```

```julia
ylabel!("Sales")
```

```julia
title!("Sales between 2000 and 2009")
```

```julia
x = -5:0.2:5
y = x.^2
```

```julia
plot(x,y)
```

```julia
p1 = plot(x, x)
p2 = plot(y, x)
p3 = plot(sin, x)
p4 = plot(cos, x)
plot(p1, p2, p3, p4)
```
