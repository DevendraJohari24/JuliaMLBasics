## Logistic Regression

```julia
using CSV
```

```julia
using DataFrames
```

```julia
dataset = CSV.read("../../JuliaDataset/camera.csv", DataFrame)
```

```julia
names(dataset)
```

```julia
using Plots
```

```julia
scatter(dataset.price, dataset.camera_mp)
```

```julia
using GLM
```

```julia
model = glm(@formula(camera_mp ~ price), dataset, Binomial(), LogitLink())
```

```julia
predict(model, DataFrame(price=[390]))
```
